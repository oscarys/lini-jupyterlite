## 1. Introduction

A **Butterworth lowpass filter** of order $N$ is characterized by a *maximally flat* magnitude response in the passband. This means the first $2N-1$ derivatives of $|H(j\Omega)|$ with respect to $\Omega$ vanish at $\Omega = 0$ — no ripple whatsoever, just a smooth monotonic roll-off.

A **normalized** Butterworth LPF is defined so that the $-3\,\text{dB}$ cutoff frequency is exactly:

$$\Omega_c = 1 \text{ rad/s}$$

Any other cutoff frequency is obtained by a simple frequency scaling of the normalized design.

The squared magnitude response is:

$$|H(j\Omega)|^2 = \frac{1}{1 + \left(\dfrac{\Omega}{\Omega_c}\right)^{2N}}$$

**Our goal** in this notebook is to derive and compute all poles of $H(s)$ in the $s$-plane for a given order $N$, verify them against `scipy.signal`, and visualize their geometry.

---
## 2. From Magnitude to Transfer Function

### 2.1 The analytic continuation trick

We relate $|H(j\Omega)|^2$ to a rational function of $s$ using the identity:

$$|H(j\Omega)|^2 = H(j\Omega)\,H^*(j\Omega) = H(s)\,H(-s)\Big|_{s\,=\,j\Omega}$$

With $\Omega_c = 1$ and substituting $s = j\Omega$ (so $\Omega = s/j = -js$):

$$H(s)\,H(-s) = \frac{1}{1 + (-s^2)^N} = \frac{1}{1 + (-1)^N s^{2N}}$$

### 2.2 Finding the 2N poles

The poles of $H(s)H(-s)$ are the $2N$ roots of the denominator:

$$1 + (-1)^N s^{2N} = 0$$

$$s^{2N} = (-1)^{N+1} = e^{\,j\pi(2m+1)}, \qquad m = 0, 1, \ldots, 2N-1$$

Taking the $2N$-th root:

$$\boxed{s_m = e^{\,j\pi(2m+1)/(2N)}}, \qquad m = 0, 1, \ldots, 2N-1$$

All $2N$ poles lie on the **unit circle** in the $s$-plane, equally spaced by $\Delta\theta = \pi/N$ radians.

---
## 3. Selecting the Stable Poles

The $2N$ poles come in **mirror-image pairs** across the imaginary axis:

- $N$ poles in the **Left Half-Plane** (LHP): $\text{Re}(s_m) < 0$
- $N$ poles in the **Right Half-Plane** (RHP): $\text{Re}(s_m) > 0$

The factorization $H(s)H(-s)$ assigns:

| | Poles |
|---|---|
| $H(s)$ — causal, stable filter | LHP poles |
| $H(-s)$ — anti-causal | RHP poles |

> **Stability criterion:** A causal LTI system is BIBO stable if and only if all poles of $H(s)$ lie in the open left half-plane.

Re-indexing the $N$ LHP poles with $k = 1, 2, \ldots, N$:

$$\boxed{s_k = e^{\,j\pi(2k + N - 1)/(2N)}}, \qquad k = 1, 2, \ldots, N$$

Expanding into real and imaginary parts:

$$\sigma_k = \cos\!\left(\frac{\pi(2k+N-1)}{2N}\right) < 0, \qquad
\omega_k = \sin\!\left(\frac{\pi(2k+N-1)}{2N}\right)$$

### Key geometric properties

1. **Unit circle:** $|s_k| = 1$ for all $k$ — all poles are exactly on the unit circle.
2. **Equal spacing:** consecutive poles are separated by $\Delta\theta = \pi/N$.
3. **Conjugate symmetry:** poles appear in conjugate pairs $s_k, s_k^*$ — ensuring real coefficients in $H(s)$.
4. **Odd order:** one pole is purely real at $s = -1$; the rest form conjugate pairs.
5. **No poles on the imaginary axis:** the filter is always strictly stable.

---
## 4. The Transfer Function

Once the poles are known, the normalized transfer function is:

$$H(s) = \frac{1}{\displaystyle\prod_{k=1}^{N}(s - s_k)}$$

For example, for $N = 2$ the two poles are at:

$$s_{1,2} = e^{\,j\cdot 3\pi/4},\; e^{\,j\cdot 5\pi/4} = -\frac{\sqrt{2}}{2} \pm j\frac{\sqrt{2}}{2}$$

Giving the classic second-order Butterworth:

$$H(s) = \frac{1}{s^2 + \sqrt{2}\,s + 1}$$

which has $Q = 1/\sqrt{2} \approx 0.707$ — the **Butterworth quality factor**.

```python
# ── Cell 1: load packages ─────────────────────────────────────────
# Run this first (Shift+Enter). Packages load inside the browser —
# no internet connection needed after this point.
print('⏳ Loading packages into the browser kernel...')
print('   (NumPy · SciPy · Matplotlib — first load takes ~10–20 s)')
print()

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from scipy.signal import buttap

%matplotlib inline
plt.rcParams.update({
    'figure.facecolor': '#fafafa',
    'axes.facecolor':   '#ffffff',
    'axes.grid':        True,
    'grid.color':       '#e8e8e8',
    'grid.linewidth':   0.6,
    'font.family':      'DejaVu Sans',
    'axes.spines.top':  False,
    'axes.spines.right':False,
})

print(f'✅ Kernel ready!')
print(f'   NumPy {np.__version__} · Matplotlib {plt.matplotlib.__version__}')
print()
print('👉 Now run Cell 2 below (Shift+Enter) to compute the poles.')
```

```python
# ── Cell 2: pole computation ──────────────────────────────────────
# ↓ Change N and re-run this cell (Shift+Enter)
N = 4

# Analytic formula: s_k = exp(jπ(2k+N-1)/(2N)), k=1..N
k  = np.arange(1, N + 1)
sk = np.exp(1j * np.pi * (2*k + N - 1) / (2*N))

# Scipy cross-check
_, poles_sp, _ = buttap(N)

# ── Print comparison table ────────────────────────────────────────
print(f'Normalized Butterworth LPF  |  Order N = {N}')
print(f'Cutoff: Ωc = 1 rad/s')
print()
print(f'{"k":>3}  {"Analytic  sₖ":^40}  {"SciPy buttap":^40}  Match')
print('─' * 95)
for i, (sa, ss) in enumerate(zip(sk, poles_sp), 1):
    match = '✓' if abs(sa - ss) < 1e-9 else f'Δ={abs(sa-ss):.2e}'
    print(f'{i:>3}  {sa:+.10f}  {ss:+.10f}  {match}')

print()
print(f'All Re(sₖ) < 0 : {np.all(sk.real < 0)}')
print(f'All |sₖ| = 1   : {np.allclose(np.abs(sk), 1.0)}')
print(f'Angular spacing: π/N = {np.degrees(np.pi/N):.4f}°  ({np.pi/N:.6f} rad)')
```

```python
# ── Cell 3: s-plane visualization ────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(13, 5.5))

# ── Left: pole-zero map ──
ax = axes[0]

# Unit circle
theta = np.linspace(0, 2*np.pi, 500)
ax.plot(np.cos(theta), np.sin(theta),
        color='#aaaaaa', lw=1.2, ls='--', label='Unit circle')

# Axes
ax.axhline(0, color='#555', lw=0.8)
ax.axvline(0, color='#555', lw=0.8)

# LHP shading
ax.axvspan(-2, 0, alpha=0.04, color='steelblue', label='LHP (stable region)')

# Analytic poles (×)
ax.scatter(sk.real, sk.imag,
           s=200, color='steelblue', marker='x',
           linewidths=2, zorder=6, label='Analytic $s_k$')

# Scipy poles (○)
ax.scatter(poles_sp.real, poles_sp.imag,
           s=80, facecolors='none', edgecolors='tomato',
           linewidths=1.5, zorder=5, label='SciPy buttap')

# Labels
for i, pole in enumerate(sk, 1):
    ax.annotate(
        f'  $s_{{{i}}}$\n  ({pole.real:.3f},{pole.imag:+.3f}j)',
        xy=(pole.real, pole.imag),
        fontsize=7, color='#333',
        va='center', linespacing=1.4
    )

ax.set_xlim(-1.7, 1.7)
ax.set_ylim(-1.7, 1.7)
ax.set_aspect('equal')
ax.set_xlabel(r'$\sigma$ — Real part', fontsize=10)
ax.set_ylabel(r'$j\omega$ — Imaginary part', fontsize=10)
ax.set_title(f'Pole locations in the s-plane  (N = {N})',
             fontsize=11, fontweight='bold', pad=12)
ax.legend(fontsize=8, loc='upper right')

# ── Right: magnitude response ──
ax2 = axes[1]

omega = np.logspace(-1, 1, 800)
H_mag = 1 / np.sqrt(1 + omega**(2*N))
H_dB  = 20 * np.log10(H_mag)

ax2.semilogx(omega, H_dB, color='steelblue', lw=2, label=f'N = {N}')
ax2.axvline(1.0, color='tomato', lw=1, ls='--', alpha=0.8,
            label=r'$\Omega_c = 1$ rad/s')
ax2.axhline(-3.0103, color='tomato', lw=0.8, ls=':', alpha=0.6)
ax2.fill_between(omega, H_dB, -80, alpha=0.08, color='steelblue')

# Asymptote annotation
ax2.annotate(f'−{20*N} dB/decade',
             xy=(4, -55), fontsize=8.5,
             color='steelblue', fontstyle='italic')

ax2.set_xlim(0.1, 10)
ax2.set_ylim(-80, 5)
ax2.set_xlabel(r'$\Omega$ (rad/s) — log scale', fontsize=10)
ax2.set_ylabel(r'$|H(j\Omega)|$ (dB)', fontsize=10)
ax2.set_title(f'Magnitude response  (N = {N})',
              fontsize=11, fontweight='bold', pad=12)
ax2.legend(fontsize=9)

plt.suptitle(f'Normalized Butterworth LPF  |  N = {N}  |  $\\Omega_c = 1$ rad/s',
             fontsize=12, y=1.01)
plt.tight_layout()
plt.show()
```

```python
# ── Cell 4: compare multiple orders side by side ──────────────────
orders  = [1, 2, 3, 4, 5, 6, 8]
colors  = plt.cm.Blues(np.linspace(0.35, 0.95, len(orders)))
omega   = np.logspace(-1, 1, 800)

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# ── Left: all pole constellations ──
ax = axes[0]
th = np.linspace(0, 2*np.pi, 400)
ax.plot(np.cos(th), np.sin(th), color='#ccc', lw=1, ls='--')
ax.axhline(0, color='#888', lw=0.6)
ax.axvline(0, color='#888', lw=0.6)

for n, c in zip(orders, colors):
    k_  = np.arange(1, n + 1)
    sk_ = np.exp(1j * np.pi * (2*k_ + n - 1) / (2*n))
    ax.scatter(sk_.real, sk_.imag,
               s=60, color=c, zorder=4, label=f'N={n}')

ax.set_xlim(-1.6, 0.3)
ax.set_ylim(-1.5, 1.5)
ax.set_aspect('equal')
ax.set_xlabel(r'$\sigma$', fontsize=10)
ax.set_ylabel(r'$j\omega$', fontsize=10)
ax.set_title('Pole constellations for N = 1…8', fontsize=11, fontweight='bold')
ax.legend(fontsize=7.5, ncol=2, loc='upper right')

# ── Right: magnitude responses ──
ax2 = axes[1]
for n, c in zip(orders, colors):
    H_dB = 20 * np.log10(1 / np.sqrt(1 + omega**(2*n)))
    ax2.semilogx(omega, H_dB, color=c, lw=1.6, label=f'N={n}')

ax2.axvline(1.0, color='tomato', lw=1, ls='--', alpha=0.7, label=r'$\Omega_c$')
ax2.axhline(-3.0103, color='tomato', lw=0.7, ls=':', alpha=0.5)
ax2.set_xlim(0.1, 10)
ax2.set_ylim(-80, 5)
ax2.set_xlabel(r'$\Omega$ (rad/s)', fontsize=10)
ax2.set_ylabel(r'$|H|$ (dB)', fontsize=10)
ax2.set_title('Magnitude responses — all orders', fontsize=11, fontweight='bold')
ax2.legend(fontsize=7.5, ncol=2)

plt.tight_layout()
plt.show()
```

## 6. Denormalization

To obtain a Butterworth LPF with arbitrary cutoff $\Omega_c$, apply a **frequency scaling** to the normalized poles:

$$s_k^{(\Omega_c)} = \Omega_c \cdot s_k$$

The normalized transfer function becomes:

$$H_{\Omega_c}(s) = H_n\!\left(\frac{s}{\Omega_c}\right)$$

All poles simply scale radially — they remain on a circle of radius $\Omega_c$ instead of the unit circle, with identical angular positions.
